### 1.1. Common Built-in Numpy Functions to Create Arrays

<p>Numpy has a number of built in methods which help us quickly and easily create multidimensional arrays.</p>

<b style='color:blue'>np.zeros</b>:

This function creates a new array filled with zeros. The size of the array is determined by the shape provided as an argument. The `np.zeros((3,2))` command, for instance, creates an array with 3 rows and 2 columns filled with zeros.

```python
import numpy as np 

zeros_array = np.zeros((3,2))      
print(zeros_array)    
```

<b style='color:blue'>np.ones</b>:

The `np.ones` function works similarly to np.zeros, but instead fills the array with ones. 

```python
ones_array = np.ones((3, 5), dtype=float)
print(ones_array)
```

<b style='color:blue'> np.full </b>:

The `np.full` function generates an array filled with the value you provide. It requires two arguments: the shape of the array and the value to fill it with. In this case, a 5x3 array is being created, filled entirely with the value 5.

```python
full_array = np.full((5, 3), 5)
print(full_array)
```

<b style='color:blue'>np.eye</b>:

The `np.eye` function is used to create a 2-D identity matrix with ones on the diagonal and zeros elsewhere. An identity matrix is a square matrix (same number of rows and columns) in which all the elements of the principal diagonal are ones and all other elements are zeros.

```python
eye_array = np.eye(5,5)
print(eye_array) 
```

### 1.2. Array Slicing (sub-array)

<p>Array slicing in NumPy allows you to extract a portion or subset of a given array. It follows a familiar syntax akin to Python lists, but with extended capabilities, especially for multi-dimensional arrays.</p>

<b>Syntax:</b>

<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<i>an_array[start:stop:step]</i></p>

Let's start with a one-dimensional array:

```python
import numpy as np

an_array = np.arange(10)  
print("Array: ", an_array)
print("Shape: ", an_array.shape)
```

Here are different sample ways you can slice this array:

```python
# Basic slicing examples for rank 1 array
slices = [an_array[2:5], 
          an_array[:5], 
          an_array[5:], 
          an_array[::2], 
          an_array[3::2], 
          an_array[::-1], 
          an_array[5::-2]]

for index, slice_data in enumerate(slices, start=1):
    print(f"Slice {index}: {slice_data}")
```

<b>Views vs. Copies in Array Slicing</b>

By default, when you slice a NumPy array, you're creating a view of the original data, not a new copy. This means changes to the sliced array can affect the original array. However, if you want a separate copy, you can create it using the `copy()` method.

```python
ex_array = np.array([[11,12,13],
                     [21,22,23],
                     [31,32,33]])


sub_array = ex_array[1:3, 1:3]


sub_array[1, 1] = 555


print("Original Array after modifying sub_array:\n", ex_array)

ex_array_copy = np.array([[11,12,13],
                          [21,22,23],
                          [31,32,33]])


sub_array_copy = ex_array_copy[1:3, 1:3].copy()

sub_array_copy[1, 1] = 555

print("\nOriginal Array after modifying sub_array_copy:\n", ex_array_copy)
```

### 1.5. Universal functions (ufunc)

Universal functions (often termed 'ufunc') in NumPy offer efficient element-wise operations applied across arrays. These functions are vectorized and much faster than their counterpart Python loops.

Consider a simple operation where we want to calculate the reciprocals of an array:

```python
np.random.seed(5)

# Using Python loop
def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

values = np.random.randint(1, 10, size=5)
print(compute_reciprocals(values))
```

```python
1/values
```

For large arrays, using loops in Python can be computationally slow. Let's see how much time it takes for a big array:

```python
big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

### 1.6. Broadcasting

<p>Broadcasting is simply a set of rules for applying binary ufuncs (addition, subtraction, multiplication, etc.) on arrays of different sizes.</p>
<img src="https://jakevdp.github.io/PythonDataScienceHandbook/figures/02.05-broadcasting.png" width=400 height=400>
<p><b>Source</b>: VanderPlas, J., 2016. Python data science handbook: essential tools for working with data. " O'Reilly Media, Inc."</p>

<p>Rules of broadcasting:</p>

* If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading side.

* If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.

* If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

```python
# Example 1
an_array = np.array([0,1,2])      # Shape (3,)
an_integer = 5
an_array + an_integer 
```

```python
# Example 2
a_matrix = np.ones((3,3))        # Shape (3,3)
an_array = np.array([0,1,2])     # Shape (3,)
a_matrix + an_array
```

```python
# Example 3
an_array = np.array([[0,1,2]])     # Shape (3,1)
another_array = np.array([[0],
                          [1],
                          [2]])  # Shape (1,3)
an_array + another_array
```

<p><b>Example:</b> (1) create a 3x4 matrix, (2) create a 3x1 vector. Add this two entities and figure out how broadcasting applies to this case?</p>

```python
# Write your code over here
A = np.random.randn(3,4)
B = np.ones((1,3))
print(A)
print(B)
A+B
```

#### 1.7.1. Multidimensional aggregates

<p>One common type of aggregation operation is an aggregate along a row or column.</p>

```python
# Example
np.random.seed(0)
an_array = np.random.randint(1,10,(4, 5))
print(an_array)
```

```python
# sum
print("Total sum:", an_array.sum())
print("Column-wise sum: ", an_array.sum(axis = 0))
print("Row-wise sum: ", an_array.sum(axis = 1))
```

```python
# mean
print("Total mean:", an_array.mean())
print("Column-wise mean: ", an_array.mean(axis = 0))
print("Row-wise mean: ", an_array.mean(axis = 1))
```

```python
# max
print("Overall max:", an_array.max())
print("Column-wise max: ", an_array.max(axis = 0))
print("Row-wise max: ", an_array.max(axis = 1))
```

<b>Note:</b> When you apply multi-dimensional operation on 2d array, the output is 1d array. Let's see an example:

```python
an_array = np.random.randint(1,10,(4, 5))
sum_result = an_array.sum(axis = 0)
sum_result
```

If you want to keep the number of dimentions in 2d format you need to use <b style="color:red">keepdims=True</b> parameter.

```python
an_array = np.random.randint(1,10,(4, 5))
sum_result = an_array.sum(axis = 0, keepdims=True)
sum_result
```

### 1.10. Merging data sets

<p>You can merge arrays horizontally or vertically, considering the following rules </p>

* In order to merge arrays vertically, two arrays should have identical number of columns.
* In order to merge arrays horizontally, two arrays should have identical number of rows.

```python
np.random.seed(0)
array_one = np.random.randint(1,10,size=(4,2))
array_two = np.random.randint(1,10,size=(3,2))
print(array_one)
print()
print(array_two)
```

```python
# vertical merge method 1
np.vstack((array_one,array_two))
```

```python
# vertical merge method 2
np.concatenate([array_one,array_two], axis = 0)
```

```python
np.random.seed(0)
array_three = np.random.randint(1,10,size=(3,4))
array_four = np.random.randint(1,10,size=(3,6))
print(array_three)
print()
print(array_four)
```

```python
# Horizontal merge method 1
np.hstack((array_three,array_four))
```

```python
# Horizontal merge method 2
np.concatenate([array_three,array_four], axis = 1)
```
